```
import matplotlib.pyplot as plt

from math import pi
 
# Set data
#df = pd.DataFrame({
#'group': ['A','B','C','D'],
#'var1': [38, 1.5, 30, 4],
#'var2': [29, 10, 9, 34],
#'var3': [8, 39, 23, 24],
#'var4': [7, 31, 33, 14],
#'var5': [28, 15, 32, 14]
#})
 
 
 
# ------- PART 1: Create background
plt.style.use('fivethirtyeight')
# number of variable
categories=list(songs100)[1:]
N = len(categories)
 
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
 
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories, size=10)
 
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=5)
plt.ylim(0,40)
 
 
# ------- PART 2: Add plots
 
# Plot each individual = each line of the data
# I don't do a loop, because plotting more than 3 groups makes the chart unreadable
 
# Ind1
values=songs100.loc[0].drop('acousticness').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="Big Bank feat. 2 Chainz, Big Sean, Nicki Mina")
ax.fill(angles, values, 'b', alpha=0.1)
 
# Ind2
values=songs100.loc[1].drop('acousticness').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label="BAND DRUM (feat. A$AP Rocky)")
ax.fill(angles, values, 'r', alpha=0.1)
 
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.01, 0.01))
```

```
def zero():
    return songs100.loc[0].drop('acousticness').values.flatten().tolist()
 
def one():
    return songs100.loc[1].drop('acousticness').values.flatten().tolist()
 
def two():
    return "two"
 
switcher = {
        0: zero,
        1: one,
        2: two
    }
 
 
def change_loc(argument):
    # Get the function from switcher dictionary
    func = switcher.get(argument, "nothing")
    # Execute the function
    return func()
 
#Input: numbers_to_strings(1)
#Output: One
 
#Input: switcher[1]=two #changing the switch case
#Input: numbers_to_strings(1)
#Output: Two
```
