```
#track id input and output
target = df_spot1['track_index_num']
feautures = df_spot1[['danceability','acousticness', 'energy','liveness','speechiness','valence']]
#popularity should be standerized
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(feautures)
print(scaler.mean_)
fea_scale = scaler.transform(feautures)
```

```
model = Sequential()

model.add(Dense(65, input_dim=6, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(130663))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```
# Training the model
model.fit(fea_scale,
          target,
          epochs=1, 
          verbose=10, 
          #validation_split=.1,
          #callbacks=callbacks
          )
```

```

# Testing the predictions of the model
model.predict(train)
```
